# UTKFace: Download and Preprocess

This notebook downloads the UTKFace dataset into curated-dataset/ and then runs the preprocessing script.

Prerequisites:
- Kaggle credentials configured for kagglehub (KAGGLE_USERNAME/KAGGLE_KEY or ~/.kaggle/kaggle.json).
- Python dependencies installed (see README for details).

```python
from pathlib import Path
import subprocess
import sys

project_dir = Path.cwd()
print(project_dir)
if not (project_dir / "download_dataset.py").exists():
    candidate = project_dir / "../"
    if candidate.exists():
        project_dir = candidate

download_script = project_dir / "download_dataset.py"
preprocess_script = project_dir / "pre_processing.py"

if not download_script.exists() or not preprocess_script.exists():
    raise FileNotFoundError("Cannot find project scripts. Run this notebook from the project_1 folder.")

image_size = 224  # Set to None to keep original size
output_dir = "pre-processed"

print("Step 1/2: Downloading dataset...")
subprocess.run([sys.executable, str(download_script)], check=True, cwd=str(project_dir))

```

## Preprocesing Dataset

```python
print("Step 2/2: Preprocessing dataset...")
preprocess_cmd = [sys.executable, str(preprocess_script), str(project_dir / "curated-dataset")]
if image_size is not None:
    preprocess_cmd.append(str(image_size))
preprocess_cmd.extend(["--output-dir", output_dir])
subprocess.run(preprocess_cmd, check=True, cwd=str(project_dir))

print("Done.")
```
